# Matplotlib

Matplotlib is a powerful plotting library in Python used for creating static, animated, and interactive visualizations. Matplotlibâ€™s primary purpose is to provide users with the tools and functionality to represent data graphically, making it easier to analyze and understand. It was originally developed by John D. Hunter in 2003 and is now maintained by a large community of developers.

Pyplot is a Matplotlib module that provides a MATLAB-like interface. Matplotlib is designed to be as usable as MATLAB, with the ability to use Python and the advantage of being free and open-source. Each pyplot function makes some changes to a figure.

```python
from matplotlib import pyplot as plt                                      # pip install matplotlib
from matplotlib import style
```

```python
#style.use('ggplot')                                   # graph style             # commented it out coz it was styling all the graphs

a = [2, 4, 5]
b = [3, 5, 7]
c = [2, 6, 9]
d = [4, 5, 8]

# first call to plt.plot creates the axes(region of the image with the data space),
# then subsequent calls to plt.plot add additional lines on the same axes

plt.plot(a,b, label='A', linewidth =3)                  # plots (2,3), (4, 5), (5, 7) on the graph
plt.plot(c, d, label='B')
plt.legend()                                            # for the legends
plt.xlabel("X-axis")                                    # names the x-axis
plt.ylabel("Y-axis")                                    # names the y-axis
plt.title('Line Graph')                                 # title
plt.show()
```

```python
import numpy as np

x = np.arange(0, 10, 0.2)
y = np.sin(x)

print("X : ", x)
print("Y : ", y)
```

Two ways to create and visualize plots

1. Explicitly create figure and axis
2. Use pyplot 

```python
fig, ax = plt.subplots()     # create a figure and an axes

ax.plot(x, y, label= 'linear')
```

```python
plt.plot(x, y, label='linear')
```

```python
plt.close()
```

```python
yy = np.cos(x)
```

```python
plt.plot(x, y, label='Sine', linewidth=4)
plt.plot(x, yy, label='Cos')
plt.plot(x, yy, label='Cos2')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Plot')
```

##### Plotting Subplots

```python
plt.subplot(2, 1, 1)         # (rows, column, plot no)
plt.plot(x, y)
plt.title("Sine Plot")

plt.subplot(2, 1, 2)
plt.plot(x, yy)
plt.title("Cos Plot")
```

## Plotting Different Graphs

### 1.  Line Graph

Line graphs are commonly used to visualize trends over time or relationships between variables.

A line plot is a type of plot that displays info as a series of data points connected by straight lines.

* Visualizing trends and changes over time
* Showing relationships
* Compare multiple data series
* Highlighting sudden changes or anomalies

First one was an example of Line Graph.

### 2.  Bar Chart

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories. 

It can be created using the bar() method.

* Also known as a bar graph
* Compares the value of variable

```python
x = [3, 1, 3, 10, 2, 4, 4] 
y = [2, 1, 1, 3, 4, 5, 6] 
plt.bar(x, y)
plt.title("Bar Chart")
plt.show()
```

### 3. Scatter Chart

Scatter plots are ideal for visualizing the relationship between two continuous variables.

* Displays values pertaining to two variables
* Determines the correlation between the two variables

```python
x = [3, 1, 2, 8, 2, 4, 6]
y = [2, 1, 1, 3, 4, 5, 3]
plt.scatter(x, y)
plt.title("Scatter chart")
plt.show()
```

### 4. Area Plot

* Also known as **Area Chart** or **Area Graph**
* Displays magnitude and proportion of multiple values
* Represents time or another ordered dimension
* Is based on the line plot


### 5. Histograms

* Represents the frequency distribution of the dataset
   - partitions numeric data into bins
   - assigns each data point in the data set
   - counts the number of data poins

### 6. Pie Chart

A circular statistical graphic, divided into segments to illustrate numerical proportions

### 7. Box Plot

A box plot is a way of statistically representing given data distribution through five main dimensions.

The five main dimensions are minimum, first quartile, median, third quartile, and maximum.


### 8. Pivot Charts

* Pivot charts are a powerful tool used for data visualization and analysis. 
* They allow users to dynamically summarize and explore large datasets, revealing insights and trends that might not be immediately obvious. 
* Pivot charts are widely used in business intelligence, finance, marketing, and various other fields where data analysis is crucial.
